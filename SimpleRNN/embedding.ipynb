```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
max_features = 10000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)
print(f"Training data shape: {X_train.shape}, Testing label shape: {y_train.shape}")
print(f"Training data shape: {X_train.shape}, Testing label shape: {y_test.shape}")
```

```python
sample_review = X_train[0]
sample_label = y_train[0]
print(f"Sample review (encoded): {sample_review}")
print(f"Sample label: {sample_label}")
```

```python
word_index = imdb.get_word_index()

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])

decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in sample_review])
print(decoded_review)
```

```python
max_len = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_len)
X_test = sequence.pad_sequences(X_test, maxlen=max_len)
print(f"Training data shape: {X_train.shape}, Testing label shape: {y_train.shape}")
print(f"Training data shape: {X_train.shape}, Testing label shape: {y_test.shape}")
X_train
```

```python
X_train[0]
```

```python
model = Sequential()
model.add(Embedding(max_features, 128, input_length=max_len))
model.add(SimpleRNN(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
early_stopping
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=10, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
model.save('sentiment_rnn_imdb.h5')
```
