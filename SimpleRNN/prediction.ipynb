```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
word_index = imdb.get_word_index()

reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
model = load_model('sentiment_rnn_imdb.h5')
model.summary()
```

```python
model.get_weights()
```

```python
def decode_review(text):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in text])

def preprocess_text(text):
    word = text.lower().split()
    word_indices = [word_index.get(w, 2) + 3 for w in word]
    return sequence.pad_sequences([word_indices], maxlen=500) 
```

```python
def predict_sentiment(text):
    processed_text = preprocess_text(text)
    prediction = model.predict(processed_text)
    sentiment ='Positive' if prediction[0][0] > 0.5 else 'Negative'
    return sentiment, prediction[0][0]
```

```python
example_review = "This movie was fantastic! I really enjoyed it."
sentiment,score = predict_sentiment(example_review)
print(f"Review: {example_review}")
print(f"Sentiment: {sentiment}")
print(f"Score: {score}")
```
